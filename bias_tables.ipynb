```python
def get_residuals(df, y, x):
    """Get residuals from OLS regression
    """
    X = df[x]
    X = sm.add_constant(X)
    model = OLS(np.log(df[y]), X)
    results = model.fit()
    return results.resid
```

```python
experiments = df_final["experiment"].unique()
metrics_dict = {"dpmi" : "PMI", "wefat_score_glovewc" : "GloVe", 
                "wefat_score_sgns": "SGNS", "wefat_score_ft" : "FastText"}

for metric in metrics_dict.keys():
    df_final[f"resid_{metric}"] = np.nan
    for experiment in experiments:
        mask_ = df_final["experiment"] == experiment
        df_final.loc[mask_, f"resid_{metric}"] = get_residuals(df_final[mask_], "score", metric)    
```

```python
# for each bias measure and experiment keep the top 20 absolute residuals
dfs_residuals = []
for experiment in experiments:
    df_exp = df_final.query("experiment == @experiment").copy()
    df_exp["metric"] = ""
    for metric, metric_name in metrics_dict.items():
        words_ = df_exp.sort_values(f"resid_{metric}", ascending=False, key=abs).head(10)["word"].values
        mask_words = df_exp["word"].isin(words_)
        df_exp.loc[mask_words, "metric"] = df_exp.loc[mask_words, "metric"] + metric_name + "; "
    dfs_residuals.append(df_exp.query("metric != ''").copy())
df_residuals = pd.concat(dfs_residuals)
# drop duplicates
df_residuals = df_residuals.drop_duplicates(subset=["experiment", "word"]).copy()
# del dfs_residuals
```

```python
INPUT_FILES_INFO = {
    "files_pmi" : [
        ("wiki2021", "gender", "results/bias_pmi-wiki2021-FEMALE-MALE.csv"),
        ("wiki2021", "valence", "results/bias_pmi-wiki2021-PLEASANT-UNPLEASANT.csv"),
        ("wiki2021", "affluence", "results/bias_pmi-wiki2021-POOR-RICH.csv"),
        ("wiki2021", "race", "results/bias_pmi-wiki2021-BLACK-WHITE.csv"),
    ],
    "files_sgns" : [
        ("wiki2021", "gender", "results/bias_sgns-wiki2021-FEMALE-MALE.csv"),
        ("wiki2021", "valence", "results/bias_sgns-wiki2021-PLEASANT-UNPLEASANT.csv"),
        ("wiki2021", "affluence", "results/bias_sgns-wiki2021-POOR-RICH.csv"),
        ("wiki2021", "race", "results/bias_sgns-wiki2021-BLACK-WHITE.csv"),
        
    ],
    "files_ft" : [
        ("wiki2021", "gender", "results/bias_ft-wiki2021-FEMALE-MALE.csv"),
        ("wiki2021", "valence", "results/bias_ft-wiki2021-PLEASANT-UNPLEASANT.csv"),
        ("wiki2021", "affluence", "results/bias_ft-wiki2021-POOR-RICH.csv"),
        ("wiki2021", "race", "results/bias_ft-wiki2021-BLACK-WHITE.csv"),
    ],
    "files_glovewc" : [
        ("wiki2021", "gender", "results/bias_glovewc-wiki2021-FEMALE-MALE.csv"),
        ("wiki2021", "valence", "results/bias_glovewc-wiki2021-PLEASANT-UNPLEASANT.csv"),
        ("wiki2021", "affluence", "results/bias_glovewc-wiki2021-POOR-RICH.csv"),
        ("wiki2021", "race", "results/bias_glovewc-wiki2021-BLACK-WHITE.csv"),
    ]
}
```

```python
dfs = {}
for k, v in INPUT_FILES_INFO.items():
    n = k.split("_")[1]
    dfs[n] = pd.concat(
        [pd.read_csv(f) for _, _, f in v], 
        keys=[(corpus, bias) for corpus, bias, _ in v],
        names=["corpus", "bias"]).reset_index(level=[0,1])

df_bias = join_bias_dfs(dfs["pmi"], dfs["sgns"],dfs["ft"],dfs["glovewc"])
del dfs
```
